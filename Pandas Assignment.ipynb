```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
x = np.where(df['FlightNumber'].isnull() == True)
for i in x:
    df['FlightNumber'][i] = df['FlightNumber'][i-1] + 10
    
df
```

```python
df['FlightNumber'] = df['FlightNumber'].astype(int)
```

```python
df
```

```python
xf = pd.DataFrame()
xf[['From','To']] = df.From_To.str.split("_",expand=True)
xf
```

```python
xf['From'] = pd.Series(xf['From']).str.title()
xf['To'] = pd.Series(xf['To']).str.title()
xf
```

```python
df.drop(columns='From_To' , axis=1 , inplace=True)
```

```python
df = pd.concat([df,xf] , axis=1)
```
